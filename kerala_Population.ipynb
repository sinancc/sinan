```python
df = pd.read_excel('../content/sample_data/Kerala_districts.xlsx')
df.head()
```

```python
merged= pd.concat([shp_gdf, df], axis=1, ignore_index=False)
merged
```

```python
fig, ax = plt.subplots(1, figsize=(12, 12))
ax.axis('off')
ax.set_title('Population District wise',
             fontdict={'fontsize': '15', 'fontweight' : '3'})
fig = merged.plot(column='Population', cmap='RdYlGn', label=f'Population',linewidth=0.5, ax=ax, edgecolor='0.2',legend=True)
```
